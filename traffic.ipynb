```python
import torch
import cv2
import numpy as np
```

```python
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)
def detect_and_track_objects(frame):
    img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    results = model(img)
    labels, cord = results.xyxyn[0][:, -1], results.xyxyn[0][:, :-1]
    
    return labels, cord

def plot_boxes(results, frame):
    labels, cord = results
    n = len(labels)
 
    height, width, _ = frame.shape

    for i in range(n):
        row = cord[i]
        if row[4] >= 0.3: 
            x1, y1, x2, y2 = int(row[0] * width), int(row[1] * height), int(row[2] * width), int(row[3] * height)
            label = f"{model.names[int(labels[i])]} {row[4]:.2f}"
            
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 255, 255), 2)
    
    return frame
```

```python

def run_object_detection_on_video(video_source="traffic 3.mp4"):

    cap = cv2.VideoCapture(video_source)

    cv2.namedWindow("Object Detection & Tracking", cv2.WINDOW_NORMAL)

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame")
            break
        results = detect_and_track_objects(frame)
       
        frame = plot_boxes(results, frame)
        
        height, width, _ = frame.shape
        cv2.resizeWindow("Object Detection & Tracking", width, height)
        
        cv2.imshow('Object Detection & Tracking', frame)
        
        if cv2.waitKey(1) & 0xFF == ord('q'):  
            break
    
    cap.release()
    cv2.destroyAllWindows()
```

```python
if __name__ == "__main__":
    run_object_detection_on_video()
  
```
